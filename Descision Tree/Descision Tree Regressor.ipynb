```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()
```

```python
print(diabetes['DESCR'])
```

```python
df = pd.DataFrame(data=diabetes['data'], columns=diabetes['feature_names'])
```

```python
df.head()
```

```python
df['target'] = diabetes['target']
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
X = df.drop('target', axis=1)
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree = DecisionTreeRegressor()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
predictions
```

```python
y_test
```

```python
plt.scatter(y_test, predictions)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score, mean_squared_log_error
```

```python
print("MAE: ", mean_absolute_error(y_test, predictions))
```

```python
print("MSE: ", mean_squared_error(y_test, predictions))
```

```python
print("RMSE: ", np.sqrt(mean_squared_error(y_test, predictions)))
```

```python
print("R2 Score: ", r2_score(y_test, predictions))
```

```python
print("Explained Variance Score: ", explained_variance_score(y_test, predictions))
```

```python
print("Mean Squared Log Error: ", mean_squared_log_error(y_test, predictions))
```

```python
df['target'].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.displot(df['target'])
```

## HPT

```python
param = {
    'criterion': ['mse', 'friedman_mse', 'mae'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5, 10, 15, 20, 25],
    'min_samples_split': [2, 5, 10, 15, 20, 25, 30],
    'min_samples_leaf': [1, 2, 5, 10, 15, 20, 25, 30],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=dtree, param_grid=param, verbose=3, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
plt.scatter(y_test, grid_predictions)
```

```python
print("MAE: ", mean_absolute_error(y_test, grid_predictions))
```

```python
print("MSE: ", mean_squared_error(y_test, grid_predictions))
```

```python
print("RMSE: ", np.sqrt(mean_squared_error(y_test, grid_predictions)))
```

```python
print("R2 Score: ", r2_score(y_test, grid_predictions))
```

```python
print("Explained Variance Score: ", explained_variance_score(y_test, grid_predictions))
```

```python
print("Mean Squared Log Error: ", mean_squared_log_error(y_test, grid_predictions))
```
