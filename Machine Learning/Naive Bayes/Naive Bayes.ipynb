```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report: \n", classification_report(y_test, y_pred))
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.info()
```

```python
category_columns = df.select_dtypes(include='category').columns
category_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoder_data = encoder.fit_transform(df[category_columns])
```

```python
encoded_columns = encoder.get_feature_names_out(category_columns)
```

```python
df[encoded_columns] = encoder_data
```

```python
df = df.drop(columns=['sex', 'smoker', 'day', 'time'])
```

```python
df
```

```python
df.columns
```

```python
X = df.drop(columns=['smoker_No', 'smoker_Yes'])
y = df['smoker_Yes']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report: \n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Smoker', 'Smoker'], yticklabels=['Not Smoker', 'Smoker'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.histplot(df[df['smoker_Yes'] == 1]['total_bill'], kde=True, color='blue', label='Smoker')
sns.histplot(df[df['smoker_Yes'] == 0]['total_bill'], kde=True, color='red', label='Non-Smoker')

plt.title('Distribution of Total Bill for Smoker vs Non-Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.legend()
plt.show()
```
